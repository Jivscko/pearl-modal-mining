## Tips

- **Stop mining**: Click Stop button in toolbar or press Ctrl+C in cell
- **Monitor GPU**: Run `!nvidia-smi -l 5` in separate cell
- **Change GPU**: Select from GPU dropdown in toolbar, then reconnect
- **Burst pattern**: Mine 1-2 hours, stop, wait, repeat (reduces detection risk)

## Cost Calculation

```python
# Example: T4 GPU at $0.60/hr
gpu_cost_per_hour = 0.60
mining_hours = 2.0
prl_earned = 0  # Fill this after mining session
prl_price_usd = 0  # Current PRL price

total_cost = gpu_cost_per_hour * mining_hours
total_revenue = prl_earned * prl_price_usd
profit = total_revenue - total_cost

print(f"Cost: ${total_cost:.2f}")
print(f"Revenue: ${total_revenue:.2f}")
print(f"Profit: ${profit:.2f} ({profit/total_cost*100:.1f}% ROI)")
```
